```python
import pandas as pd

# Load and skip metadata/header rows (first 2 rows)
aqi_meaning = pd.read_csv("aqi_meaning.csv", skiprows=2)
air_quality_index = pd.read_csv("air_quality_index.csv", skiprows=2)
air_purifier = pd.read_csv("air_purifier.csv", skiprows=2)
pm_25 = pd.read_csv("pm_2.5.csv", skiprows=2)

# Rename columns
aqi_meaning.columns = ['Week', 'AQI Meaning']
air_quality_index.columns = ['Week', 'Air Quality Index']
air_purifier.columns = ['Week', 'Air Purifier']
pm_25.columns = ['Week', 'PM 2.5']

# Merge all on 'Week'
merged = aqi_meaning.merge(air_quality_index, on='Week', how='outer')
merged = merged.merge(air_purifier, on='Week', how='outer')
merged = merged.merge(pm_25, on='Week', how='outer')

# Convert to datetime and sort
merged['Week'] = pd.to_datetime(merged['Week'])
merged = merged.sort_values('Week')

# Export to CSV
merged.to_csv("GoogleTrends_AQI_Awareness.csv", index=False)

# Preview
merged.head()
```
