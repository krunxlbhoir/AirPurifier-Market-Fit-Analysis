```python
import pandas as pd

# Load datasets
aqi_df = pd.read_csv('/content/aqi.csv')
pop_df = pd.read_csv('/content/population_projection.csv')

# Step 1: Filter population data
# Use only 'Total' gender and year 2024
pop_filtered = pop_df[(pop_df['year'] == 2024) & (pop_df['gender'] == 'Total')]

# Convert value from thousands to actual population
pop_filtered['urban_population'] = pop_filtered['value'] * 1000

# Get total urban population per state (in case a state appears multiple times across months)
pop_state = (
    pop_filtered.groupby('state')['urban_population']
    .mean()  # or use sum() if appropriate
    .reset_index()
)

# Step 2: Prepare AQI Data
aqi_df['date'] = pd.to_datetime(aqi_df['date'], errors='coerce')
aqi_2024 = aqi_df[aqi_df['date'].dt.year == 2024]

aqi_state_avg = (
    aqi_2024.groupby('state')['aqi_value']
    .mean()
    .reset_index()
    .rename(columns={'aqi_value': 'avg_aqi'})
)

# Step 3: Merge the two datasets
merged_df = pd.merge(pop_state, aqi_state_avg, on='state', how='inner')

# Step 4: Export
merged_df.to_csv('State_Population_vs_AQI.csv', index=False)

# Preview
merged_df.head()
```
