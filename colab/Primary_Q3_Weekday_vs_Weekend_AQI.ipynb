```python
import pandas as pd

# Load the dataset
aqi_df = pd.read_csv('aqi.csv')

# Convert to datetime
aqi_df['date'] = pd.to_datetime(aqi_df['date'], format="%d-%m-%Y", errors='coerce')

# Step 1: Filter for last 1 year (July 2024 â€“ July 2025)
aqi_1yr = aqi_df[(aqi_df['date'] >= '2024-07-01') & (aqi_df['date'] <= '2025-07-01')]

# Step 2: Filter for selected metro cities
metro_cities = ['Delhi', 'Mumbai', 'Chennai', 'Kolkata', 'Bengaluru', 'Hyderabad', 'Ahmedabad', 'Pune']
aqi_metros = aqi_1yr[aqi_1yr['area'].isin(metro_cities)]

# Step 3: Add weekday/weekend column
aqi_metros['day_type'] = aqi_metros['date'].dt.dayofweek.apply(lambda x: 'Weekend' if x >= 5 else 'Weekday')

# Step 4: Calculate average AQI by city and day type
weekend_vs_weekday = (
    aqi_metros.groupby(['area', 'day_type'])['aqi_value']
    .mean()
    .reset_index()
    .rename(columns={'aqi_value': 'avg_aqi'})
)

# Step 5: Save to CSV
weekend_vs_weekday.to_csv("MetroCities_WeekdayWeekend_AQI.csv", index=False)

# Display result
weekend_vs_weekday
```
